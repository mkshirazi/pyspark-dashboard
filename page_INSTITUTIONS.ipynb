```python
# Main Tables Used for visuals

# Main authors data
authors_data = data.select("id","year",explode("authors").alias("authors"))
authors_data = authors_data.select(authors_data.id.alias("paper_id"),\
    F.year("year").alias("year"),\
    F.col("authors").getItem("name").alias("name"), \
    F.col("authors").getItem("country").alias("country"), \
    F.col("authors").getItem("id").alias("author_id"), \
    F.col("authors").getItem("affiliation").alias("affiliation"), \
    F.col("authors").getItem("affiliationid").alias("affiliationid"), \
    F.col("authors").getItem("order").alias("order"))

authors_data.cache()

# Topics per Inst
inst_topics = authors_data.join(data.select("id","topics"),authors_data.paper_id==data.id,"left")\
            .select("affiliation",explode("topics").alias("topics"))\
            .groupBy(["affiliation","topics"]).count()

# Author Stats
# Review g-index and fix h-index
from pyspark.sql.window import Window
windowSpec  = Window.partitionBy("author_id").orderBy(col("citationcount").desc())
authors_stats = authors_data.join(data.select("id","citationcount","topics"),authors_data.paper_id==data.id,"left")\
                            .withColumn("h_index",F.when(F.dense_rank().over(windowSpec)<=col("citationcount"),1).otherwise(0))\
                            .groupBy("author_id")\
                            .agg(F.first("name").alias("name"),\
                                F.first("country").alias("country"),\
                                F.first("affiliation").alias("affiliation"),\
                                F.first("affiliationid").alias("affiliationid"),\
                                F.count("paper_id").alias("papers_count"),\
                                F.avg("order").alias("avg_order"),\
                                F.avg("citationcount").alias("avg_citations"),\
                                F.sum("citationcount").alias("total_citations"),\
                                F.sum("h_index").alias("h_index"),\
                                F.first("paper_id").alias("last_paper_id"),\
                                F.collect_set("topics").alias("all_topics"))\
                            .sort("h_index",ascending=False)

authors_stats.cache()

# Inst Stats
inst_stats = authors_data.groupBy("affiliationid")\
                         .agg(F.first("affiliation").alias("name"),\
                            F.first("country").alias("country"),
                            F.countDistinct("paper_id").alias("papers_count"),\
                            F.countDistinct("author_id").alias("authors_count"))
                          
inst_stats.cache()

# Coauthors
coinst_data = authors_data.select("paper_id","author_id","affiliation","affiliationid")\
                              .join(data.select("id",explode("authors").alias("authors")),authors_data.paper_id==data.id,"left")\
                              .select("paper_id", "affiliation","affiliationid",\
                                      F.col("authors").getItem("affiliation").alias("co_inst"))\
                              .groupBy(["affiliationid","affiliation","co_inst"]).agg(F.countDistinct("paper_id").alias("co_inst_count"))

coinst_data.cache()
```

## MAIN DASHBOARD

```python
app = JupyterDash(external_stylesheets=[dbc.themes.DARKLY])

app.layout = html.Div([ 
          dbc.Container([
             dbc.Row([
                  dbc.Col(
                          html.H3(children = "Institutions in CS Research", 
                                  style = {'color':'pink', 'align':'center'},
                                  className="text-center mt-4 mb-5"), align="center", className="mt4")       
             ]),
              dbc.Row([
                  dbc.Col(
                          html.Div(children = [
                              html.Label("Institute", className="text-center mt-4 mb-5",
                                       style = {'width': '85vw', "height":'0.01vw', 'align':'center', 'padding':0, 'margin':0}),
                              dcc.Dropdown(id = "inst-select",
                                        options = [{'label': i, 'value': i} for i in institutions_list["affiliation"].sort_values().tolist()],
                                        value = "University of Waterloo",
                                        style={"display": "inline-block",
                                            "width": "65%",
                                            "margin-left": "2px",
                                            "verticalAlign": "top",
                                            "color":"black"}
                                        )],className="text-center" ),
                          width = 12,
                          align="center")
              ]),
              html.Br(),
              html.Br(),
              dbc.Row([
                  dcc.Graph(id="time-chart")
              ]),
              dbc.Row([
                  dcc.Graph(id='papers-table')
                  ]),
              html.Br(),
              dbc.Row([
                  dbc.Col(
                      html.Div(
                      dcc.Graph(id="topics-table")), align="center", width = 5
                  ),
                  dbc.Col(
                      html.Div(
                      dcc.Graph(id="coinst-chart")), align="center", width = 7
                  )
              ])
          ])
])



@app.callback(Output(component_id="time-chart", component_property="figure"),
               Input(component_id="inst-select", component_property="value"))
def update_time_chart(inst_input):
    df = authors_data.filter(col("affiliation")==inst_input).\
                      filter(col("year")>=2010).\
                      groupBy("year").agg(F.countDistinct("paper_id").alias("Number of Papers")).orderBy("year").toPandas()
    df.set_index("year", drop=True, inplace=True)
    fig = px.line(df, color_discrete_sequence = ["#ff1493"], range_x=[2010,2020])
    fig.layout.template = "plotly_dark"
    title = f"Papers trends for {inst_input}"
    fig.update_layout(plot_bgcolor='rgba(0, 0, 0, 0)',paper_bgcolor='rgba(0, 0, 0, 0)', title=title, title_x =0.5)
    return fig



@app.callback(Output(component_id="papers-table", component_property="figure"),
               Input(component_id="inst-select", component_property="value"))
def update_table(inst_input):
    top_authors = authors_stats.select("name","papers_count","h_index","avg_citations","total_citations","all_topics")\
                        .filter(col("affiliation")==inst_input)\
                        .limit(10).toPandas()
    top_authors.name =  top_authors.name.apply(lambda x: x.title())
    top_authors.avg_citations = top_authors.avg_citations.apply(round)
    top_authors.all_topics = top_authors.all_topics.apply(lambda lst: list(set(x for l in lst for x in l))[:10])#lambda y: list(set(sum(y,[]))))

    fig = go.Figure(data=[go.Table(
          columnorder = [1,2,3,4,5,6],
          columnwidth = [80,80,80,80,80,300],
          header=dict(values=["Name","Number of Papers","H Index","Avg. Citations","Total Citations Count","Top Topics"],
                      fill_color='black',
                      align='left',
                      font=dict(color='white', size=14)),
          cells=dict(values=[top_authors.name,top_authors.papers_count,top_authors.h_index,round(top_authors.avg_citations,2),\
                            top_authors.total_citations,top_authors.all_topics.apply(lambda x: ', '.join(map(str, x)))],
                    fill_color='lavender',
                    align='left',
                    font=dict(color='black', size=14)))
      ])
    fig.layout.template = "plotly_dark"
    title = f"Top 10 Authors from {inst_input}"
    fig.update_layout(plot_bgcolor='rgba(0, 0, 0, 0)',paper_bgcolor='rgba(0, 0, 0, 0)',title=title, title_x=0.5,
                     height=600)
    return fig


@app.callback(Output(component_id="topics-table", component_property="figure"),
               Input(component_id="inst-select", component_property="value"))
def update_table(inst_input):
    topics = inst_topics.filter(col("affiliation")==inst_input)\
                .select("topics","count")\
                .sort("count",ascending=False).limit(20).toPandas()

    fig = go.Figure(data=[go.Table(
          columnorder = [1,2],
          columnwidth = [100,50],
          header=dict(values=["Topic","Number of Papers"],
                      fill_color='black',
                      align='left',
                      font=dict(color='white', size=14)),
          cells=dict(values=[topics["topics"].apply(lambda x: x.title()),topics["count"]],
                    fill_color='#ffe4e1',
                    align='left',
                    font=dict(color='#505050', size=14)))
                    ])
    fig.layout.template = "plotly_dark"
    title = f"Popular Topics at {inst_input}"
    fig.update_layout(plot_bgcolor='rgba(0, 0, 0, 0)',paper_bgcolor='rgba(0, 0, 0, 0)',title=title, title_x=0.5,
                      margin=dict(l=20, r=20, t=30, b=20) )
    return fig

@app.callback(Output(component_id="coinst-chart", component_property="figure"),
               Input(component_id="inst-select", component_property="value"))
def update_pie(inst_input):
    co_insts = coinst_data.filter(col("affiliation")==inst_input)\
                            .select("co_inst","co_inst_count")\
                            .sort("co_inst_count",ascending=False)\
                            .toPandas()[1:]
    co_insts = co_insts.dropna()
    df2 = co_insts[:8].copy()
    new_row = pd.DataFrame(data = {
          'co_inst' : ['others'],
          'co_inst_count' : [co_insts['co_inst_count'][8:].sum()]
      })

    co_insts = pd.concat([df2, new_row])
    co_insts.rename({'co_inst': 'Co-Affiliation', 'co_inst_count': 'Number of Papers'}, axis=1, inplace=True)

    fig = px.pie(co_insts, values="Number of Papers", names='Co-Affiliation', color='Co-Affiliation')
    fig.update_traces(textposition='inside', textinfo='label',hoverinfo='label+percent+value')
      # fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
    fig.layout.template = "plotly_dark"
    title = f"Top Co-Affiliations"
    fig.update_layout(plot_bgcolor='rgba(0, 0, 0, 0)',paper_bgcolor='rgba(0, 0, 0, 0)',title=title, title_x=0.5,
                      margin=dict(l=20, r=20, t=30, b=20) )
    return fig
    


app.run_server(mode="external", port=8085)
```

```python
institution_name = "University of Waterloo"
top_author_names = authors_stats.select("name","all_topics")\
                        .filter(col("affiliation")==institution_name)\
                        .limit(10)

top_topics = top_author_names.select(explode("all_topics").alias("topic")).select(explode("topic").alias("topic")).groupBy("topic").agg(F.count("topic").alias("count")).sort("count", ascending = False).limit(10)
author_topic = authors_data.filter("affiliation = \'" + institution_name + "\'").select("paper_id", "name", "affiliation")\
    .join(data.select("id"),authors_data.paper_id==data.id,"left").join(top_author_names, on = "name", how="inner")\
    .select("name", "affiliation", explode("all_topics").alias("topic")).\
    select("name", "affiliation", explode("topic").alias("topic")).join(top_topics, on="topic", how = "inner")

author_topic.show()

df = author_topic.toPandas()
fig = px.parallel_categories(df, dimensions=['affiliation', 'name', 'topic'],
                labels={'affiliation':'Affiliation', 'name':'Author', 'topic':'Topic'})
fig.show()
```
